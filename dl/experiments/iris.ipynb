````python
# Pour la manipulation de tableaux et Dataframes
import numpy as np
import pandas as pd

# Pour la visualisation des performances
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Pour instancier une couche Dense et modГЁle sГ©quentiel
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

from pathlib import Path
DATA_DIR = Path("/app/data")
````

````python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# Les colonnes ne sont pas nommГ©es dans le fichier brut de l'UCI
colnames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris_df = pd.read_csv(url, names=colnames)
# Sauvegarde dans un fichier CSV
# iris_df.to_csv(str(DATA_DIR / "iris_data.csv"), index=False)
iris_df.head()
````

````python
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
````

````python
X = iris_df.drop(columns='species')
y = iris_df['species']
````

````python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=1/3, random_state=42)
````

LвҖҷentraГ®nement dвҖҷun rГ©seau de neurones se dГ©roule, pour faire simple, en trois Г©tapes principales :

##### Etape 1 : DГ©finition de lвҖҷ```architecture du modГЁle```
	nombre de couche et de neurones et comment les donnГ©es passeront Г  travers le rГ©seau

##### Etape 2 : ```Compilation``` du modГЁle
	- la fonction de perte => la qualitГ© de la prГ©diction (mГ©sure l'erreur entre la pred et le rГ©el)
	- mГ©triques => Г©valuer les performances
	- algorithme d'optimisation => ajuster les poids pour rГ©duire l'erreur

##### Etape 3 : ```EntraГ®nement``` du modГЁle : utiliser la mГ©thode fit

Ces Г©tapes constituent le cЕ“ur du processus dвҖҷentraГ®nement dвҖҷun rГ©seau de neurones.

#### Etape 1 : DГ©finition de lвҖҷarchitecture du modГЁle

````python
# CrГ©ation du modГЁle sГ©quentiel
model = Sequential()

# On dГ©finit d'abord la forme de l'entrГ©e
model.add(Input(shape=(4,)))

# Ajout des couches dans l'ordre
model.add(Dense(units=10, activation="tanh"))
model.add(Dense(units=8, activation="tanh"))
model.add(Dense(units=6, activation="tanh"))
model.add(Dense(units=3, activation="softmax"))
````

````python
## Affichez la structure du modГЁle en utilisant la mГ©thode summary
model.summary()
# On peut observer le nombre de paramГЁtres que le modГЁle doit apprendre, 
# c'est-Г -dire le nombre de poids qu'il doit ajuster lors de l'entraГ®nement Г  l'aide de la descente de gradient.
````

On peut observer le nombre de paramГЁtres que le modГЁle doit apprendre, c'est-Г -dire le nombre de poids qu'il doit ajuster lors de l'entraГ®nement Г  l'aide de la descente de gradient.
#### D'oГ№ viennent les chiffres dans Param # ?
La formule mathГ©matique pour calculer le nombre de paramГЁtres d'une couche Dense est : 
``ParamГЁtres = (entrГ©es x nb_neurones) + nb_neurones``
<br> exemple : 
     <ul>
     <li>50 = (4 x 10) + 10</li>
     <li>88 = (10 x 8) + 8</li>
     <li>54 = (8 x 6) + 6</li>
     <li>21 = (6 x 3) + 3</li>
    </ul>     

#### Etape 2 : Compilation du modГЁle

````python
model.compile(loss="sparse_categorical_crossentropy",
             optimizer="adam",
             metrics=["accuracy"])
````

#### 1. loss="sparse_categorical_crossentropy" (La fonction de perte)
    <ul>
        <li>Pourquoi "Categorical Crossentropy" ? Parce que tu fais de la classification</li>
        <li>Pourquoi "Sparse" ? tes Г©tiquettes (les rГ©ponses attendues dans tes donnГ©es) sont de simples nombres entiers (par exemple : 0 pour "Chat", 1 pour "Chien", 2 pour "Oiseau"). Si tu n'avais pas mis sparse, tu aurais dГ» encoder tes rГ©ponses sous forme de vecteurs (ex: [1, 0, 0]), ce qu'on appelle le One-Hot Encoding.</li>
    </ul>


#### Etape 3 : EntraГ®nement du modГЁle

````python
history = model.fit(X_train, y_train, epochs=500, batch_size=32, validation_split=0.1)
````

#### validation_split=0.1 (Le test surprise)

C'est une excellente pratique ! Tu demandes Г  Keras de mettre de cГҙtГ© 10 % de tes donnГ©es d'entraГ®nement et de ne surtout pas s'entraГ®ner dessus.

    ГҖ la fin de chaque Г©poque, le modГЁle va passer un "examen" sur ces 10 % de donnГ©es cachГ©es.

    Cela va te donner deux types de scores dans ta console : l'accuracy sur les donnГ©es d'entraГ®nement (accuracy) et l'accuracy sur les donnГ©es de validation (val_accuracy).

    Si ton accuracy monte Г  98% mais que ta val_accuracy stagne Г  60%, c'est le signe que ton modГЁle apprend par cЕ“ur (overfitting).

#### 1. DГ©cryptage des scores (Гүpoque 500)

    accuracy: 0.9889 (98,89 %) : Sur tes donnГ©es d'entraГ®nement, le modГЁle classe correctement prГЁs de 99 % des exemples. Il ne se trompe presque jamais.

    loss: 0.0622 : L'erreur globale est trГЁs proche de 0. C'est le signe que l'optimiseur a fait du trГЁs bon travail.

    val_accuracy: 1.0000 (100 %) : C'est le score parfait. Sur les donnГ©es de validation (les donnГ©es "surprises" qu'il n'a pas vues pendant l'entraГ®nement), le modГЁle a fait un sans-faute.

    val_loss: 0.0239 : La perte de validation est extrГӘmement basse, encore plus basse que celle d'entraГ®nement, ce qui confirme que le modГЁle est trГЁs stable.

#### 2. Le verdict : Pas de surapprentissage (Overfitting) !

C'Г©tait le piГЁge avec 500 Г©poques, mais ton modГЁle l'a brillamment Г©vitГ©. On parle de surapprentissage lorsque l'accuracy d'entraГ®nement est excellente (ex: 99 %) mais que l'accuracy de validation s'effondre (ex: 70 %).

Ici, les deux scores (accuracy et val_accuracy) sont trГЁs proches et trГЁs hauts. Ton modГЁle n'a pas bГӘtement appris tes donnГ©es par cЕ“ur, il a rГ©ellement compris la logique gГ©nГ©rale de ton problГЁme.
